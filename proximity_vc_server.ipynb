```python
# 1️⃣ 既存の node index.js を止める（特定の index.js に限定）
!ps aux | grep '[n]ode index.js' | awk '{print $2}' | xargs -r kill -9

# 2️⃣ Node.js をバックグラウンドで再起動、PID を記録
import subprocess

# Node.jsをバックグラウンドで起動（標準出力・エラーはserver.logにリダイレクト）
with open("server.log", "w") as log_file:
    process = subprocess.Popen(
        ["node", "index.js"],
        stdout=log_file,
        stderr=log_file
    )

print(f"Node.js started with PID {process.pid}")

!echo $! > node.pid

import re
import threading
import time
import requests # HTTPリクエスト用ライブラリ
import os

# --- json5 ライブラリのインストール (コメント付きJSON対応のため) ---
print("starting...")
install_json5_proc = subprocess.run(['pip', 'install', 'json5'], capture_output=True, text=True, check=False)
import json5 as json # 標準のjsonの代わりにjson5をjsonという名前でインポート


# --- SSHトンネル用関数 (tcpexposer.com) - 2トンネル対応 ---
def run_ssh_tunnels_with_password(config, ssh_password_from_secret):
    ssh_password = ssh_password_from_secret
    password_source_msg = "Colab Secrets"

    if not ssh_password:
        if 'ssh_password' in config and str(config.get('ssh_password', '')).strip():
            ssh_password = str(config.get('ssh_password', '')).strip()
            password_source_msg = "config.json (パスワードの平文保存は非推奨)"
        else:
            print("エラー: SSHパスワードがColab Secretsにもconfig.jsonにも有効な値で設定されていません。")
            return False

    username = str(config.get('username', '')).strip()
    if not username:
        print("エラー: SSHトンネルに必要な 'username' がconfig.jsonに設定されていません。")
        return False

    # sshpassのインストール
    install_update = subprocess.run(['apt-get', 'update'], capture_output=True, text=True, check=False)
    if install_update.returncode != 0: print(f"apt-get updateに失敗: {install_update.stderr}")
    install_sshpass = subprocess.run(['apt-get', 'install', '-y', 'sshpass'], capture_output=True, text=True, check=False)
    if install_sshpass.returncode != 0:
        print(f"sshpassのインストールに失敗: {install_sshpass.stderr}")
        print("SSHトンネルの実行を中止します。")
        return False

    tunnel_configs = []
    # トンネル1の設定 (sub_domain, web_port)
    sub_domain1 = str(config.get('sub_domain', '')).strip()
    web_port1 = str(config.get('web_port', '')).strip()
    if sub_domain1 and web_port1:
        tunnel_configs.append({
            "name": f"RoomID: {sub_domain1}\nURL：https://proximity-vc-mcbe.pages.dev?roomid={sub_domain1}",
            "sub_domain": sub_domain1,
            "local_port": web_port1
        })
    elif sub_domain1 or web_port1:
        print(f"警告: SSHトンネル1の設定が不完全です (sub_domain='{sub_domain1}', web_port='{web_port1}')。このトンネルはスキップされます。")

    # トンネル2の設定 (sub_domain2, port)
    sub_domain2_val = str(config.get('sub_domain2', '')).strip()
    port_val = str(config.get('port', '')).strip() # config.jsonの'port'を2つ目のトンネルのローカルポートとして使用
    if sub_domain2_val and port_val:
        tunnel_configs.append({
            "name": f"\nこのコマンドを実行して接続：/connect {sub_domain2_val}.tcpexposer.com\nRun this command to connect: /connect {sub_domain2_val}.tcpexposer.com",
            "sub_domain": sub_domain2_val,
            "local_port": port_val
        })
    elif sub_domain2_val or port_val:
        print(f"警告: SSHトンネル2の設定が不完全です (sub_domain2='{sub_domain2_val}', port='{port_val}')。このトンネルはスキップされます。")

    if not tunnel_configs:
        print("設定可能なSSHトンネルがありません。")
        return False

    active_ssh_processes = []

    for tc in tunnel_configs:
        ssh_command = [
            'sshpass', '-p', ssh_password, 'ssh',
            '-R', f'{tc["sub_domain"]}:80:localhost:{tc["local_port"]}',
            f'{username}@tcpexposer.com',
            '-N', # リモートコマンドを実行しない
            '-o', 'StrictHostKeyChecking=no',
            '-o', 'UserKnownHostsFile=/dev/null',
            '-o', 'ExitOnForwardFailure=no', # ローカルポートエラーでも接続維持
            '-o', 'ServerAliveInterval=60',  # 接続維持
            '-o', 'ConnectTimeout=10'       # 接続試行のタイムアウト
        ]

        masked_command_ssh = list(ssh_command)
        if masked_command_ssh[0] == 'sshpass' and masked_command_ssh[1] == '-p':
            masked_command_ssh[2] = "'********'"

        try:
            # stderrもキャプチャするため、stdoutとstderrを分けるか、マージする
            # ここではPopenのデフォルト（パイプしない）にして、エラーはコンソールに出るようにするのも一手
            # または、非同期に読み出すスレッドを立てるなど。
            # シンプルにするため、ここではエラー出力をリアルタイムでは細かく追わない。
            proc = subprocess.Popen(ssh_command, text=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            active_ssh_processes.append({'process': proc, 'name': tc['name']})
            print(f"{tc['name']}")
        except Exception as e:
            print(f"{tc['name']} の起動に失敗しました: {e}")

    if not active_ssh_processes:
        print("起動できたSSHトンネルプロセスがありません。")
        return False

    print("✅ start")

    try:
        while True:
            all_terminated = True
            for p_info in active_ssh_processes:
                proc = p_info['process']
                if proc.poll() is None: # プロセスがまだ実行中
                    all_terminated = False
                else:
                    # プロセスが終了した場合の処理
                    stdout, stderr = proc.communicate() # 終了しているのでブロックしない
                    if stdout: print(f"  [{p_info['name']} STDOUT]:\n{stdout.strip()}")
                    if stderr: print(f"  [{p_info['name']} STDERR]:\n{stderr.strip()}")
                    # 終了したプロセスをリストから実質的に除外する（監視対象から外す）
                    active_ssh_processes = [p for p in active_ssh_processes if p['process'].pid != proc.pid]
                    if not active_ssh_processes : all_terminated = True # 監視対象がなくなればループを抜ける準備

            if all_terminated and not active_ssh_processes: # 監視対象がなくなり、すべて終了処理済み
                break
            if not active_ssh_processes and tunnel_configs: # 最初から起動プロセスがなかった場合など
                 if not any(p_info['process'].poll() is None for p_info in active_ssh_processes): #念のため確認
                    print("アクティブなSSHトンネルプロセスがなくなりました。")
                    break

            if not active_ssh_processes: # もしリストが空になったらループ終了
                print("監視対象のSSHトンネルがなくなりました。")
                break

            time.sleep(5)
    except KeyboardInterrupt:
        print("\nCtrl+C が検出されました。SSHトンネルを終了します...")
    finally:
        for p_info in active_ssh_processes: # まだ監視リストに残っている（＝実行中だった）プロセス
            proc = p_info['process']
            if proc.poll() is None:
                proc.terminate()
        time.sleep(1) # terminate後少し待つ
        for p_info in active_ssh_processes:
            proc = p_info['process']
            if proc.poll() is None:
                proc.kill()
        print('\n\033[31m' + "終了しました\n再度起動する場合は最初から実行しなおしてください\nFinished\nTo start the program again, rerun it from the beginning." +'\033[0m')
    return True


# --- Cloudflareトンネル用関数 ---
def run_cloudflared_tunnels(config): # config を引数として受け取る
    CLOUDFLARE_BASE_DOMAIN = "trycloudflare.com"
    cloudflared_command = "cloudflared"

    try:
        version_proc = subprocess.run([cloudflared_command, '--version'], check=True, capture_output=True, text=True)
    except (subprocess.CalledProcessError, FileNotFoundError):
        print('\n\033[31m' + "エラーが発生しました\n最初から実行しなおしてください\nAn error has occurred\nPlease run it again from the beginning." +'\033[0m')
        return False

    # config.json からポートを読み込む (存在しない/空の場合はデフォルト値を使用)
    default_cf_port1 = "19132"
    default_cf_port2 = "8000"

    cf_port1_from_config = str(config.get('port', default_cf_port1)).strip()
    cloudflare_port1 = cf_port1_from_config if cf_port1_from_config else default_cf_port1

    cf_port2_from_config = str(config.get('web_port', default_cf_port2)).strip()
    cloudflare_port2 = cf_port2_from_config if cf_port2_from_config else default_cf_port2


    proc1, proc2 = None, None
    try:
        proc1 = subprocess.Popen(
            [cloudflared_command, 'tunnel', '--url', f'localhost:{cloudflare_port1}', '--no-autoupdate'],
            stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, universal_newlines=True
        )
        proc2 = subprocess.Popen(
            [cloudflared_command, 'tunnel', '--url', f'localhost:{cloudflare_port2}', '--no-autoupdate'],
            stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, universal_newlines=True
        )
    except FileNotFoundError:
        print('\n\033[31m' + "エラーが発生しました\n最初から実行しなおしてください\nAn error has occurred\nPlease run it again from the beginning." +'\033[0m')
        return False
    except Exception as e:
        print('\n\033[31m' + "エラーが発生しました\n最初から実行しなおしてください\nAn error has occurred\nPlease run it again from the beginning." +'\033[0m')
        return False

    # (wait_for_tunnel_host_id, send_keep_alive, 以降のロジックは前回と同様)
    def wait_for_tunnel_host_id(proc, tunnel_name):
        if proc is None or proc.stdout is None:
            print(f"[{tunnel_name}] プロセスまたはstdoutが初期化されていません。")
            return None
        # タイムアウトを追加（例: 30秒間、約100行程度）
        timeout_seconds = 30
        start_time = time.time()
        line_count = 0
        max_lines = 100 # あまりに多くの出力を防ぐ

        for line in proc.stdout:
            if time.time() - start_time > timeout_seconds or line_count > max_lines:
                print(f"[{tunnel_name}] URL取得タイムアウトまたは行数上限到達。")
                break
            line_stripped = line.strip()
            match = re.search(r'https://([a-zA-Z0-9-]+)\.trycloudflare\.com', line_stripped)
            if match:
                host_id = match.group(1)
                full_url = match.group(0)
                return host_id
            if "Failed to establish tunnel" in line_stripped or "failed to connect to origin" in line_stripped:
                print(f"[{tunnel_name}] トンネル確立失敗の兆候: {line_stripped}")
                # return None # ここでNoneを返すと、もう片方のトンネルも待たずに進む場合がある。状況による。
            line_count += 1
        print(f"[{tunnel_name}] 出力からトンネルホストIDを見つけられませんでした。")
        return None

    def send_keep_alive(host_id, base_domain, tunnel_name, interval_seconds=180):
        if not host_id:
            print(f"[{tunnel_name}] ホストIDが無効なため、キープアライブをスキップします。")
            return
        url = f"https://{host_id}.{base_domain}"
        while True:
            try:
                response = requests.head(url, timeout=15)
                current_time = time.strftime("%Y-%m-%d %H:%M:%S")
                # print(f"[{tunnel_name}] キープアライブ成功: {url} - Status: {response.status_code} ({current_time})") # ログが多いのでコメントアウトも検討
            except requests.exceptions.RequestException as e:
                current_time = time.strftime("%Y-%m-%d %H:%M:%S")
                # print(f"[{tunnel_name}] キープアライブ失敗 (RequestException): {url} - {e} ({current_time})")
            except Exception as e:
                current_time = time.strftime("%Y-%m-%d %H:%M:%S")
                # print(f"[{tunnel_name}] キープアライブ中に予期せぬエラー: {url} - {e} ({current_time})")
            time.sleep(interval_seconds)

    host_id1 = wait_for_tunnel_host_id(proc1, f"CF_Tunnel_1 (localhost:{cloudflare_port1})")
    host_id2 = wait_for_tunnel_host_id(proc2, f"CF_Tunnel_2 (localhost:{cloudflare_port2})")

    keep_alive_threads = []
    if host_id1:
        full_url1_display = f"https://{host_id1}.{CLOUDFLARE_BASE_DOMAIN}"
        print(f"\nこのコマンドを実行して接続：/connect {host_id1}.trycloudflare.com\nRun this command to connect: /connect {host_id1}.trycloudflare.com")
        thread1 = threading.Thread(target=send_keep_alive, args=(host_id1, CLOUDFLARE_BASE_DOMAIN, f"CF_Tunnel_1_KeepAlive({cloudflare_port1})"), daemon=True)
        keep_alive_threads.append(thread1)
        thread1.start()
    else:
        print(f"Cloudflareトンネル1 (localhost:{cloudflare_port1}) のURL取得に失敗しました。")

    if host_id2:
        full_url2_display = f"https://{host_id2}.{CLOUDFLARE_BASE_DOMAIN}"
        print(f"RoomID：{host_id2}")
        print(f"URL：https://proximity-vc-mcbe.pages.dev?roomid={host_id2}")
        thread2 = threading.Thread(target=send_keep_alive, args=(host_id2, CLOUDFLARE_BASE_DOMAIN, f"CF_Tunnel_2_KeepAlive({cloudflare_port2})"), daemon=True)
        keep_alive_threads.append(thread2)
        thread2.start()
    else:
        print(f"Cloudflareトンネル2 (localhost:{cloudflare_port2}) のURL取得に失敗しました。")

    if not host_id1 and not host_id2:
        print("両方のCloudflareトンネルURLの取得に失敗したため、プログラムを終了します。")
        if proc1 and proc1.poll() is None: proc1.terminate()
        if proc2 and proc2.poll() is None: proc2.terminate()
        time.sleep(0.5);
        if proc1 and proc1.poll() is None: proc1.kill()
        if proc2 and proc2.poll() is None: proc2.kill()
        return False

    print("✅ start")
    try:
        while True:
            p1_active = proc1 and proc1.poll() is None and (host_id1 or not any(msg in str(proc1.stdout.peek() if proc1.stdout else b'') for msg in ["Failed", "failed"])) # URL取得済みか、まだ失敗してない場合
            p2_active = proc2 and proc2.poll() is None and (host_id2 or not any(msg in str(proc2.stdout.peek() if proc2.stdout else b'') for msg in ["Failed", "failed"]))

            if not (p1_active or p2_active) and (host_id1 or host_id2): # URL取得成功したトンネルが全て終了
                 print("アクティブなCloudflareトンネルプロセスがなくなりました。")
                 break
            if not (host_id1 or host_id2) and not (p1_active and p2_active) : # URL取得に両方失敗し、プロセスも両方終了
                 print("URL取得に失敗し、Cloudflareプロセスも終了しました。")
                 break

            if keep_alive_threads and not any(t.is_alive() for t in keep_alive_threads):
                print("全てのキープアライブスレッドが終了しました。メインループを終了します。")
                break

            time.sleep(5)
    except KeyboardInterrupt:
        print("\nCtrl+C が検出されました。Cloudflareトンネルを終了します...")
    finally:
        procs_to_kill = []
        if proc1 and proc1.poll() is None: procs_to_kill.append(proc1)
        if proc2 and proc2.poll() is None: procs_to_kill.append(proc2)
        for p in procs_to_kill: p.terminate()
        time.sleep(1)
        for p in procs_to_kill:
            if p.poll() is None: p.kill()
        print('\n\033[31m' + "終了しました\n再度起動する場合は最初から実行しなおしてください。\nFinished\nTo start the program again, rerun it from the beginning." +'\033[0m')
    return True


# --- メイン処理 ---
def main():
    config = {}
    ssh_password_from_secret = ""
    use_ssh_tunnel_flag = False

    try:
        with open('config.json', 'r', encoding='utf-8') as f:
            config = json.load(f)
    except FileNotFoundError:
        print('\n\033[31m' + "'config.json' が見つかりません\n‘Config‘を実行してください" + '\033[0m')
        print("'config.json' not found\nPlease run ‘Config’")
    except json.JSONDecodeError as e:
        print('\n\033[31m' + f"'config.json' の読み込みに失敗しました (JSON形式エラー)\nConfigを正しく記述してください\nFailed to read 'config.json' (JSON format error) \nPlease write Config correctly" + '\033[0m')
        print(e)
    except Exception as e:
        print('\n\033[31m' + f"config.jsonの処理中に予期せぬエラーが発生しました\n最初からやり直してくださいください\nAn unexpected error occurred while processing config.json \nPlease start over from the beginning." + '\033[0m')
        print(e)

    username_in_config = str(config.get('username', '')).strip()
    # SSHトンネル1用の設定
    sub_domain1_in_config = str(config.get('sub_domain', '')).strip()
    web_port1_in_config = str(config.get('web_port', '')).strip()
    # SSHトンネル2用の設定
    sub_domain2_in_config = str(config.get('sub_domain2', '')).strip()
    port_in_config = str(config.get('port', '')).strip() # 'port'キーを使用

    ssh_password_in_config = str(config.get('ssh_password', '')).strip()
    has_ssh_password = bool(ssh_password_from_secret or ssh_password_in_config)

    # SSHトンネル方式を選択する基本条件
    can_attempt_ssh_tunnel = username_in_config and has_ssh_password
    # 実際に起動できるSSHトンネルがあるか (少なくとも1つのsub_domainとportのペアが有効か)
    has_valid_ssh_tunnel_config = (sub_domain1_in_config and web_port1_in_config) or \
                                  (sub_domain2_in_config and port_in_config)

    if can_attempt_ssh_tunnel and has_valid_ssh_tunnel_config:
        use_ssh_tunnel_flag = True
    else:
        print("\nConfig.jsonにTCPexposerの設定がありません\nランダムなドメインで実行されます\nThere is no TCPexposer setting in Config.json\nRuns in a random domain")


    if use_ssh_tunnel_flag:
        run_ssh_tunnels_with_password(config, ssh_password_from_secret)
    else:
        run_cloudflared_tunnels(config) # configを渡す

if __name__ == '__main__':
    main()
```
